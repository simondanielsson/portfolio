```python
import pandas as pd
import numpy as np
import sklearn
```

# 1. Raw data exporation

```python
path = "./train_features.csv"

data = pd.read_csv(path)
```

```python
data.describe()
```

```python
data.info()
```

```python
# number
missing=data.isnull().sum().reset_index().rename(columns={0:'missNum'})
# percentage
missing['missRate']=missing['missNum']/data.shape[0]
# order
miss_analy=missing[missing.missRate>0].sort_values(by='missRate',ascending=False)
```

```python
miss_analy
```

```python
data = data.sort_values(by=["pid","Time"])
```

# 2. Transform into new format: one row for each patient

```python
data_1hours = []
data_p = pd.DataFrame()
for i in range(12):
    id_ith = range(i,len(data),12)
    data_1hour = data.iloc[id_ith]
    columns = data_1hour.columns
    columns_new = ["pid"]
    for j in range(1,len(columns)):
        col_new = columns[j] + "-" + str(i)
        columns_new.append(col_new)
    
    data_1hour.columns = columns_new
    
    data_1hours.append(data_1hour)
    
    if data_p.empty:
        data_p = data_1hour
    else:
        data_p = pd.merge(data_p,data_1hour,how="inner",on="pid") 
```

```python
# write into csv
data_p_path = "./data_p.csv"
data_p.to_csv(data_p_path, header=True, index=False)
```

```python
data_miss = pd.DataFrame()
for i in range(12):
    id_ith = range(36*i+1,36*(i+1))
    miss_1hour = missing_p.iloc[id_ith]
    columns = missing_p.columns
    columns_new = ["index"]
    col_new = columns[1] + "-" + str(i)
    columns_new.append(col_new)
    
    miss_1hour.columns = columns_new
    miss_1hour = miss_1hour.reset_index(drop=True)
    
    if data_miss.empty:
        data_miss = miss_1hour
    else:
        data_miss = pd.merge(data_miss,miss_1hour,left_index=True,right_index=True) 
```
